```python
# checking for missing data
missing_entries = df.isnull().sum()
missing_entries
```

```python
# ok now that it's established there is no missing data, i'll edit the csv file programatically to be more accurate
# Update Atalanta's Wins
df.loc[df["Team"] == "Atalanta", "Wins"] = 11
df.loc[df["Team"] == "Atalanta", "Draws"] = 1
df.loc[df["Team"] == "Atalanta", "Losses"] = 3

# Recalculate Points for Atalanta (using 3 points per win + 1 point per draw)
df.loc[df["Team"] == "Atalanta", "Points"] = (
    df.loc[df["Team"] == "Atalanta", "Wins"] * 3 +
    df.loc[df["Team"] == "Atalanta", "Draws"]
)
```
